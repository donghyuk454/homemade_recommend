```python
import pandas as pd
import numpy as np
import scipy as sp
from sklearn.metrics.pairwise import cosine_similarity
import operator
%matplotlib inline
```

```python
ratings = pd.read_csv('./sample_data/rating.csv', encoding='utf-8')
recipes = pd.read_csv('./sample_data/recipe.csv', encoding='utf-8')
recipes_ratings = pd.merge(ratings, recipes, on='recipe_id')
recipes_ratings.head()
```

```python
recipe_usr_rating = recipes_ratings.pivot_table('rating', index='recipe', columns='user_id')
usr_recipe_rating = recipes_ratings.pivot_table('rating', index='user_id', columns='recipe')
```

```python
recipe_usr_rating.head()
```

```python
recipe_usr_rating.fillna(0, inplace=True)
recipe_usr_rating.head()
```

```python
similarity_rate = cosine_similarity(recipe_usr_rating, recipe_usr_rating)
print(similarity_rate)
```

```python
similarity_rate_df = pd.DataFrame(
    data=similarity_rate,
    index=recipe_usr_rating.index,
    columns=recipe_usr_rating.index
)

similarity_rate_df.head()
```

```python
np.save('./result.npy', similarity_rate_df.to_numpy())
```
